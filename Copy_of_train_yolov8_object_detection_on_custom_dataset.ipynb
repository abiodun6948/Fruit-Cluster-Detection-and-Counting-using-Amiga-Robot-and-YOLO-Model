```python
import glob
from IPython.display import Image, display

# Define the base path where the folders are located
base_path = '/content/runs/detect/'

# List all directories that start with 'predict' in the base path
subfolders = [os.path.join(base_path, d) for d in os.listdir(base_path)
              if os.path.isdir(os.path.join(base_path, d)) and d.startswith('predict25')]

# Find the latest folder by modification time
latest_folder = max(subfolders, key=os.path.getmtime)

image_paths = glob.glob(f'{latest_folder}/*.jpg')[:3]

# Display each image
for image_path in image_paths:
    display(Image(filename=image_path, width=600))
    print("\n")
```

```python
# prompt: zip and download all the images inside the predict folder

import zipfile
import os

def zip_and_download_images(folder_path, zip_filename="Day_only_car_rough.zip"):
  """Zips all images in a folder and initiates a download."""
  try:
    with zipfile.ZipFile(zip_filename, 'w') as zipf:
      for root, _, files in os.walk(folder_path):
        for file in files:
          if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif')):
            file_path = os.path.join(root, file)
            zipf.write(file_path, arcname=os.path.relpath(file_path, folder_path))
    print(f"Successfully created {zip_filename}")

    # For Google Colab, use files.download to trigger the download
    from google.colab import files
    files.download(zip_filename)
  except FileNotFoundError:
    print(f"Error: Folder '{folder_path}' not found.")
  except Exception as e:
    print(f"An error occurred: {e}")


# Example usage: Assuming your images are in 'runs/detect/predict'
zip_and_download_images('/content/runs/detect/predict25')
```
